```python
# 데이터로더 생성을 위한 라이브러리 불러오기
from mmcv.utils import Config
from mmseg.datasets import build_dataset, build_dataloader
import matplotlib.pyplot as plt
```

```python
# config file 불러오기 _ datasets 부분
config_file = '/opt/ml/level2_semanticsegmentation_cv-level2-cv-03/CV03/configs/Augmentation/Aug_Nor.py'

# config를 cfg로 관리 & 데이터셋 생성
cfg = Config.fromfile(config_file)
datasets = build_dataset(cfg.data.train)

# 초기 4개 이미지를 mask와 함께 한번에 관찰
fig = plt.figure(figsize=(16, 32))
for idx, val in enumerate(datasets):
    ax1 = fig.add_subplot(4, 2, 2*idx+1)
    img = val['img']._data.permute(1, 2, 0) * 255
    ax1.imshow(img)
    
    ax2 = fig.add_subplot(4, 2, 2*idx+2)
    mask = val['gt_semantic_seg']._data.permute(1, 2, 0) * 255
    ax2.imshow(mask)
    
    if idx == 3:
        break

plt.show()

    
```
